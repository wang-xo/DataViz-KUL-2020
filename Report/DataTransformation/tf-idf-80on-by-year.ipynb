```python
import pandas as pd
import math
import numpy as np
df = pd.read_csv("../merged_data/words_songs_matched_80on.csv", sep=",")
```

TFIDF Style Metric

```python
df.head()
```

```python
#max frequency by year
for year in df.year.unique():
    print(year)
    df.loc[df['year'] == year, 'max_f_by_year'] = df.loc[df['year'] == year].groupby('word')['count'].transform(sum).max()
df.head()
```

```python
#frequency by year
df['f_by_year'] = df.groupby(['word','year'])['count'].transform(sum)
df.head()
```

```python
#tf by year
df['tf_by_year'] = df['f_by_year']/df['max_f_by_year']
print('tf domain:', [df['tf_by_year'].min(),df['tf_by_year'].max()])
```

```python
#track_count by year
for year in df.year.unique():
    print(year)
    df.loc[df['year'] == year, 'track_count_by_year'] = df.loc[df['year'] == year]['track_id'].nunique()
df.head()
```

```python
#df by year
df['df_by_year'] = df.groupby(['word','year'])['word'].transform('count')
df.head()
```

```python
#idf
df['idf_by_year'] = np.log2(df['track_count_by_year']/df['df_by_year'])
```

```python
df['tfidf_by_year'] = df['tf_by_year']*df['idf_by_year']
print('tfidf domain:', [df['tfidf_by_year'].min(),df['tfidf_by_year'].max()])
```
