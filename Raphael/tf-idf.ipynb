```python
import pandas as pd
import math
df = pd.read_csv("../merged_data/1500k.csv", sep=",")
```

```python
df
```

```python
df[['word','count']].groupby('word').agg(sum).reset_index().sort_values(['count'],ascending=False)
```

```python
df.groupby('word')['count'].transform(sum).max()
```

```python
df['max_count_in_track'] = df.groupby('track_id')['count'].transform(max)
df['tf'] = df['count']/df['max_count_in_track']
df['df'] = df.groupby(['word'])['word'].transform('count')
```

```python
track_count = len(df['track_id'].unique())
print(track_count)
def calIDF(df, track_count):
    return math.log2(track_count/df)
df['idf'] = df['df'].apply(calIDF, track_count=track_count)
```

```python
df['tfidf'] = df['tf']*df['idf']
```

```python
print([df['tf'].max(),df['tf'].min()])
print([df['idf'].max(),df['idf'].min()])
```

```python
df.drop(columns=['song_id', 'artist_id','duration','artist_hotttnesss','year','title','artist_name'])
```

```python
df.to_csv("../merged_data/1500k_tfidf.csv", sep=",")
```
